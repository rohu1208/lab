# Implement Agglomrative hierarchy
Agglomerative hierarchical clustering is a "bottom-up" approach to hierarchical clustering, where each data point starts as its own cluster, and pairs of clusters are merged together as you move up the hierarchy.

```python
#step 1 -- import libraries 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#step 2 -- generate data using make_blobs
X,y=make_blobs(n_samples=100,centers=3,cluster_std=1.05,random_state=42)
```

```python
# X,y
```

```python
# step3 -- 3. Visualize Data
plt.scatter(X[:,0],X[:,1],c='blue',marker='o',edgecolor='black',s=50)
plt.title("Generated Data")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python
# Step 4 --> Apply Agglomerative Cust
agglomarative_clustering=AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
labels=agglomarative_clustering.fit_predict(X)
```

```python
# Step 5 --> visualize Clustering result
# Plot clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, marker='o', edgecolor='black', s=50, cmap='viridis')
plt.title("Agglomerative Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
#Step 6(Optional) Compute the linkage matrix
linked = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import numpy as np

# Sample data
X2 = np.array([[1, 2], [1, 4], [1, 0],
              [3, 2], [4, 4], [4, 0]])

# Apply Agglomerative Clustering
clustering = AgglomerativeClustering(n_clusters=2).fit(X2)

# Plotting the data points with cluster labels
plt.figure(figsize=(8, 6))
plt.scatter(X2[:, 0], X2[:, 1], c=clustering.labels_, cmap='viridis', s=100)
plt.title('Agglomerative Clustering')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.show()

linked2 = linkage(X2, method='ward')
```

```python
# optional
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked2, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```
